**Data Overview**

```python
df['laptop_id'].nunique()
df.shape
```

**Insight**

This step helps understand the number of unique records and overall size of the dataset.

**Column segregation is required:**

```python
df['name'] = df['name'].astype(str)

# Brand & Model
df['brand'] = df['name'].str.split().str[0]
df['model'] = df['name'].str.split().str[1]

# RAM
df['ram'] = df['name'].str.extract(r'(\d+)\s?GB').astype(float)

# Storage
df['storage'] = df['name'].str.extract(r'(\d+)\s?(GB|TB)')[0]
df['storage_unit'] = df['name'].str.extract(r'(\d+)\s?(GB|TB)')[1]

# Processor
df['processor'] = df['name'].str.extract(r'(i3|i5|i7|Ryzen\s?\d|N\d{4}|J\d{4})')

# OS
df['os'] = df['name'].str.extract(r'(Windows\s?\d+|Mac|DOS|Linux)')

# Display
df['display'] = df['name'].str.extract(r'(\d+\.?\d*)\s?inch')


df.head()
```

**Insight**

Key features like brand, RAM, storage, processor, OS, and display were extracted from the product name column to enable better analysis.

**Column name rename if required**

```python
df.rename(columns={
    'laptop_name': 'name',
    'price_before_discount': 'original_price',
    'delivery_to_nairobi': 'delivery_location',
    'seller_customer_rating': 'customer_rating'
}, inplace=True)
```

**Insight**

* Column names were renamed to improve clarity and consistency.

**Remove duplicates if required**

```python
df = df.drop_duplicates()
```

**Insight**

No duplicate records were found in the dataset.

**Data Cleaning if required**

```python
# Clean price column

df['price'] = df['price'].astype(str).str.replace('KSh', '', regex=False)\
                                       .str.replace(',', '', regex=False)\
                                       .str.strip()

# Clean discount price column

df['original_price'] = df['original_price'].astype(str)\
    .str.replace('KSh', '', regex=False)\
    .str.replace(',', '', regex=False)\
    .str.strip()
```

**Insight:**

Removed currency symbols and formatting issues to prepare data for numerical conversion.

**Convert data types if needed**

**1. Convert Numerical Columns**

```python
# Convert numerical columns
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['original_price'] = pd.to_numeric(df['original_price'], errors='coerce')
df['storage'] = pd.to_numeric(df['storage'], errors='coerce')
df['display'] = pd.to_numeric(df['display'], errors='coerce')
```

**Insight:**

* Converted price and hardware-related columns into numeric format for analysis.

* Invalid or improperly formatted values were converted to NaN using errors='coerce'.

**2. Convert Categorical Columns**

```python
# Convert categorical columns

cat_cols = ['brand', 'model', 'processor', 'storage_unit', 'delivery_location', 'ram']

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype(str)

        # 🔹 Special handling for RAM
        if col == 'ram':
            df[col] = df[col].str.extract('(\d+)')
            df[col] = df[col] + 'GB'

        df[col] = df[col].astype('category')

# Optional column
if 'os' in df.columns:
    df['os'] = df['os'].astype(str)
    df['os'] = df['os'].astype('category')
```

**Insight:**

Converted text-based columns into categorical type for better memory efficiency and structured analysis.

**3. Customer Rating Handling**

```python
df['customer_rating'] = df['customer_rating'].astype('category')
```

**Insight:**

* Customer rating contains multiple qualitative categories such as Very Poor, Poor, Average, Good, and Excellent, along with some missing values. Therefore, it was treated as a categorical variable.

* Encoding is deferred to the feature engineering stage to preserve the original data.

**Handle missing values (impute or drop)**

```python
# 🔹 Numerical columns (median) → removed RAM
num_cols = ['price', 'storage', 'display']

for col in num_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())


# 🔹 Handle RAM separately (categorical)
if 'ram' in df.columns:
    df['ram'] = df['ram'].astype(str)
    df['ram'] = df['ram'].str.extract('(\d+)')
    df['ram'] = df['ram'] + 'GB'
    df['ram'] = df['ram'].fillna(df['ram'].mode()[0])
    df['ram'] = df['ram'].astype('category')


# 🔹 Fix original_price using price
if 'original_price' in df.columns:
    df['original_price'] = df['original_price'].fillna(df['price'])


# 🔹 Categorical columns
cat_cols = ['brand', 'model', 'processor', 'storage_unit', 'delivery_location']

for col in cat_cols:
    if col in df.columns:
        if str(df[col].dtype) == 'category':
            if 'Unknown' not in df[col].cat.categories:
                df[col] = df[col].cat.add_categories('Unknown')
        df[col] = df[col].fillna('Unknown')


# 🔹 Customer Rating (separate handling)
if 'customer_rating' in df.columns:
    if str(df['customer_rating'].dtype) == 'category':
        if 'Unknown' not in df['customer_rating'].cat.categories:
            df['customer_rating'] = df['customer_rating'].cat.add_categories('Unknown')
    df['customer_rating'] = df['customer_rating'].fillna('Unknown')


# 🔹 OS column (separate handling)
if 'os' in df.columns:
    if str(df['os'].dtype) == 'category':
        if 'Windows' not in df['os'].cat.categories:
            df['os'] = df['os'].cat.add_categories('Windows')
    df['os'] = df['os'].fillna('Windows')
```

**Insight:**
* Median was used for price to reduce the impact of extreme values (outliers).

* Missing discount prices were replaced with actual price assuming no discount.

* Median imputation was used for numerical columns to handle skewed distributions effectively.

* Missing categorical values were replaced with Unknown to maintain dataset completeness.

* Missing processor details were labeled as Unknown to avoid data loss.

* Missing ratings were filled with Unknown to retain all records for analysis.

* Missing OS values were filled with the most common assumption (Windows).

**Treat outliers if required**

**Outlier Detection and Treatment**

**Objective**

* To identify and handle extreme values in the dataset to improve the accuracy of analysis.

```python
# Step 1: Detect Outliers using IQR

# Calculate Q1 and Q3

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

# Calculate IQR

IQR = Q3 - Q1

# Define bounds

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

lower, upper
```

```python
# Step 2: Apply Clipping (Best Method)

# Apply clipping

df['price'] = df['price'].clip(lower, upper)
```

**Insight**:

Outliers were detected using the IQR method and handled using clipping to limit extreme values without removing data.

**Check skewness and apply transformations**

**Objective**

* To identify skewness in numerical data and apply transformation to improve data distribution.

```python
# Step 1: Check Skewness

skew_value = df['price'].skew()
```

```python
# Step 2: Apply Transformation

if skew_value > 1:
    df['price'] = np.log1p(df['price'])
```

**Insight**

Skewness of the price column was checked and log transformation was applied to reduce skewness and improve data distribution.

**Dropping Unnecessary Columns**

```python
df = df.drop(columns=['scraped_at'], errors='ignore')
```

**Insight**

* The scraped_at column was removed as it contained limited time-based data (only a few days), making it unsuitable for trend analysis.

* Removing it helps simplify the dataset and focus on relevant features.

**Feature transformations (date parts, derived fields if required for analysis)**

```python
# 1. Value for Money

#  Map customer rating to numeric score
rating_map = {
    'Very Poor': 1,
    'Poor': 2,
    'Average': 3,
    'Good': 4,
    'Excellent': 5
}

df['rating_score'] = df['customer_rating'].map(rating_map)


#  Handle missing/unknown ratings
df['rating_score'] = df['rating_score'].fillna(1)   # safest fallback


#  Create Value for Money feature
df['value_for_money'] = df['price'] / df['rating_score']
```

**Insight**:

This feature helps identify products that provide better performance relative to their price.

```python
# 2. Price Difference

df['price_diff'] = df['original_price'] - df['price']
```

**Insight**:

This shows the difference between actual and discounted price, helping analyze discount patterns.

```python
# 🔹 Price Range (Low, Mid, High - balanced for BI dashboard)

df['price_range'] = pd.cut(
    df['price'],
    bins=[0, 20000, 35000, float('inf')],
    labels=['Low', 'Mid', 'High']
)
```

**Insight**:

Grouping products into price categories makes comparisons and visualization easier.
